```python
class Node:
    def __init__(self,x):
        self.value = x
        self.next = None

global head
head = None

def insert_node(position, value):
    # @param position, an integer
    # @param value, an integer
    global head
    if head is None and position != 1:
        return
    else:
        if position == 1:
            if head is None:
                head = Node(value)
            else:
                temp = Node(value)
                temp.next = head
                head = temp
        else:
            start = head
            prev = None
            count = 1
            while start is not None and count != position:
                prev = start
                start = start.next
                count = count + 1
            if start is None and count == position:
                prev.next = Node(value)
            elif start is not None and count == position:
                temp = Node(value)
                prev.next = temp
                temp.next = start
            else:
                return


def delete_node(position):
    # @param position, integer
    # @return an integer
    global head
    if head is None:
        return
    else:
        if position == 1:
            head = head.next
        else:
            start = head
            prev = None
            count = 1
            while start is not None and count != position:
                prev = start
                start = start.next
                count = count + 1
            
            if start == None:
                return
            else:
                prev.next = start.next
                del start

def print_ll():
    # Output each element followed by a space
    global head
    if head is None:
        return
    else:
        start = head
        
        while start is not None:
            print(start.value, end = ' ')
            start = start.next

        print('')
        
insert_node(1,30)
insert_node(2,40)
insert_node(3,50)
insert_node(4,60)
delete_node(3)

print_ll()
```
